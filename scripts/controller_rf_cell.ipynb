```python
# get rf estimator from pkl
import pickle
from sklearn.tree import _tree

rf_serialization_balanced_equal_path = f"../outputs/serialization/balanced/rf_balanced_equal_{flow_length}_n_estimators_3_max_depth_5_seralized.pkl"

with open(rf_serialization_balanced_equal_path, "rb") as f:
    data_plane_estimator = pickle.load(f)

leaf_index_list = []
for i, dt in enumerate(data_plane_estimator.estimators_):
    dt_structure = dt.tree_
    leaf_index_list_temp = []

    def traversal_tree(node_index, tree_index):
        if dt_structure.feature[node_index] != _tree.TREE_UNDEFINED:
            left_child_index = int(
                dt_structure.children_left[node_index])
            right_child_index = int(
                dt_structure.children_right[node_index])
            traversal_tree(left_child_index, i+1)
            traversal_tree(right_child_index, i+1)
        else:
            # all node index has been increased with 1
            leaf_index_list_temp.append(node_index+1)
    # start traversing the tree
    traversal_tree(0, i+1)
    # add the leaf indexes of this tree to the entire list
    leaf_index_list.append(leaf_index_list_temp)
```

```python
################################################ calculate macro metrics ################################################
# benign confusion matrix
fn_benign = flow_fp_p4_sum
fp_benign = flow_fn_p4_sum
tn_benign = flow_tp_p4_sum
tp_benign = flow_tn_p4_sum

# attack confusion matrix (== original)
fn_attack = flow_fn_p4_sum
fp_attack = flow_fp_p4_sum
tn_attack = flow_tn_p4_sum
tp_attack = flow_tp_p4_sum

pre_benign = tp_benign / (tp_benign + fp_benign)
re_benign = tp_benign / (tp_benign + fn_benign)
f1_benign = (2 * pre_benign * re_benign) / (pre_benign + re_benign)

pre_attack = tp_attack / (tp_attack + fp_attack)
re_attack = tp_attack / (tp_attack + fn_attack)
f1_attack = (2 * pre_attack * re_attack) / (pre_attack + re_attack)

macro_pre = (pre_benign + pre_attack) / 2
macro_re = (re_benign + re_attack) / 2
macro_f1 = (f1_benign + f1_attack) / 2

print(f"- benign precision: {pre_benign}")
print(f"- benign recall: {re_benign}")
print(f"- benign f1 score: {f1_benign}")
print(f"- attack precision: {pre_attack}")
print(f"- attack recall: {re_attack}")
print(f"- attack f1 score: {f1_attack}")
print(f"- macro precision: {macro_pre}")
print(f"- macro recall: {macro_re}")
print(f"- macro f1 score: {macro_f1}")
```
